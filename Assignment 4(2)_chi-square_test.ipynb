# CHI-SQUARE TEST

```python
import numpy as np
import math
from scipy.stats import chi2
```

```python
#H0: Device type and customer satisfaction are independent.
#H1: Device type and customer satisfaction are associated.
```

```python
alpha = 0.05 
df = 4 #degree of freedom
critical_value = 9.488
```

```python
# Chi_square value inside table is 9.488;
# So X^2 is greater than or we fail to reject the H0
```

```python
observed = np.array([
    [50,70],
    [80,100],
    [60,90],
    [30,50],
    [20,50],
])
n_rows, n_cols = observed.shape
grand_total = observed.sum()
print("Observed_value:",observed)
```

```python
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
```

```python
expected = np.outer(row_totals, col_totals)/grand_total
```

```python
chi_square = (observed - expected) **2/expected
chi_sq_stat = chi_square.sum()

print("Chi_square:\n", chi_square)
print("Sum of all Chi_square_stats:\n",chi_sq_stat)
```

```python
df = (n_rows - 1) * (n_cols - 1)
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)
```

```python
print(df)
print(alpha)
```

```python
if chi_sq_stat < critical_value:
    print("Do not reject H0 (no association)")
else:
    print("Reject H0 (association exists)")
```
