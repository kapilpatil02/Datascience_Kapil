## Confidence Interval

```python
import numpy as np
from scipy import stats
```

```python
data = (1.13,1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29)
```

```python
n = len(data)
```

```python
head_mean = np.mean(data)
```

```python
print(head_mean)
```

```python
samp_std = np.std(data, ddof=1)
```

```python
print(samp_std)
```

```python
# a. Build 99% Confidence Interval Using Sample Standard Deviation
```

```python
alpha = 0.01
```

```python
t_distribution = stats.t.ppf(1-alpha/2, df = n-1)
t_set = samp_std/np.sqrt(n)
margin_t = t_distribution*t_set
ci_t = (head_mean - margin_t, head_mean + margin_t)
```

```python
print(ci_t[0])
print(ci_t[1])
```

```python
# b.Build 99% Confidence Interval Using Known Population Standard Deviation
```

```python
pop_std = 0.20
z_distribution = stats.t.ppf(1-alpha/2, df= n-1)
z_set = pop_std/np.sqrt(n)
margin_z = z_distribution*z_set
ci_z = (head_mean - margin_z, head_mean + margin_z)
```

```python
print(ci_z[0])
print(ci_z[1])
```

```python
print(f"Sample Mean: {head_mean} ")
print(f"Sample Std: {samp_std} ")
print(f"99% C.I Using t-distribution: {ci_t[0]} to {ci_t[1]} ")
print(f"99% C.I Using z-distribution: {ci_z[0]} to {ci_z[1]} ")
```
