## Prime Numbers

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):  # check odd divisors up to sqrt(n)
        if n % i == 0:
            return False
    return True

number = int(input("Enter a number: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

## Product of Random Numbers

```python
import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print(f"What is {num1} × {num2}?")
user_answer = int(input("Enter your answer: "))

correct_answer = num1 * num2

if user_answer == correct_answer:
    print("✅ Correct! Well done.")
else:
    print(f"❌ Incorrect. The correct answer is {correct_answer}.")
```

## Squares of Even/Odd Numbers

```python
print("Squares of odd numbers from 100 to 200:\n")

for number in range(100, 201):
    if number % 2 != 0: 
        square = number ** 2
        print(f"{number}² = {square}")
```

## Word counter

```python
from collections import Counter

def word_counter(text):
    # Split the text into words based on whitespace
    words = text.split()
    
    # Count the frequency of each wordS
    word_count = Counter(words)
    
    # Print results
    for word, count in word_count.items():
        print(f"'{word}': {count}")

# Example usage
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
word_counter(input_text)
```

## Check for Palindrome

```python
import re

def is_palindrome(text: str) -> bool:
    # Remove all non-alphanumeric characters and convert to lowercase
    cleaned = re.sub(r'[^A-Za-z0-9]', '', text).lower()
    # Check if cleaned string is equal to its reverse
    return cleaned == cleaned[::-1]

# Take input from user
user_input = input("Enter a word, phrase, or number: ")

# Decide and print result
if is_palindrome(user_input):
    print("✅ Yes, it's a palindrome!")
else:
    print("❌ No, it's not a palindrome.")
```
