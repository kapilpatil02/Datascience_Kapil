```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('Alphabets_data.csv')
df.head()
```

```python
print('Shape:', df.shape)
df.info()
```

```python
df.isnull().sum().sum()
```

```python
X = df.drop('letter', axis=1)
y = df['letter']
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(len(np.unique(y_encoded)), activation='softmax'))
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
y_pred = np.argmax(model.predict(X_test), axis=1)


print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
tuned_model = Sequential()
tuned_model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
tuned_model.add(Dense(64, activation='relu'))
tuned_model.add(Dense(len(np.unique(y_encoded)), activation='softmax'))


optimizer = Adam(learning_rate=0.0005)


tuned_model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
tuned_model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=30,
    batch_size=32,
    verbose=1
)
```

```python
y_pred_tuned = np.argmax(tuned_model.predict(X_test), axis=1)

print('Tuned Accuracy:', accuracy_score(y_test, y_pred_tuned))
print(classification_report(y_test, y_pred_tuned, zero_division=0))
```
